```python
import os
from os import path
from wordcloud import WordCloud
```

```python
text = "NyaySaathi is a groundbreaking legal service e-marketplace designed to revolutionize access to justice in India. In a country where the legal service sector is largely unorganized, NyaySevaHub offers a centralized, user-friendly platform connecting individuals with a diverse range of legal service providers, including lawyers, arbitrators, mediators, notaries, and document writers. The platform's unique selling points include tailored legal matching, verified professionals, transparent pricing, and seamless remote consultations. NyaySevaHub prioritizes inclusivity, collaborating with legal aid organizations and government agencies to provide free or subsidized legal services to low-income individuals, fostering social justice. With gamification elements, robust user support, and a commitment to data security and privacy, NyaySevaHub aims to empower citizens from all backgrounds to access quality legal services, bridging the justice gap and creating a more equitable legal landscape in India."
```

```python
import matplotlib.pyplot as plt
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
```
